#### Importing libraries 

```python
from keras.models import Sequential 
from keras.layers import Dense, LSTM, Embedding
from keras.preprocessing.text import Tokenizer
from keras.utils import to_categorical
import numpy as np
```

#### Raw data

```python
data = "rain rain go away come again another day little johny wants to play johny wants to play again in the rain."
```

#### First step is the dataset preparation and to do that we need to first tokenize our data. 

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts([data])
```

#### now our data is tokenized and we need to print the indexes 

```python
sequence = tokenizer.texts_to_sequences([data])[0]
sequence
```

#### Here we'll calculate the number of indexes we have 

```python
token = len(tokenizer.word_index) + 1
token
```

#### we need to make bigram pairs as we are makeing a model for predicting the next word 

```python
sequences = []
for i in range(1, len(sequence)):
    seq = sequence[i-1: i+1]
    print(seq)
    sequences.append(seq)
sequences = np.array(sequences)
```

#### let's separate out these into train_x and train_y
 

```python
train_x, train_y = sequences[:, 0], sequences[:, 1]
train_y = to_categorical(train_y, num_classes = token)
```

#### Defining model architecture 

```python
model = Sequential() #initializing sequential model

model.add(Embedding(token, 10, input_length = 1))
model.add(LSTM(32))
model.add(Dense(token, activation = "softmax"))

model.compile(loss = "categorical_crossentropy", optimizer = "adam", metrics = ["accuracy"])
model.summary()
```

#### Fitting the dataset on model 

```python
model.fit(train_x, train_y, epochs = 500, verbose = 1)
```

#### Now our model is trained, lets see how it can be used to generate text 

```python
def generate_text(input_text):
    result = input_text
    
    #we'll take the range of 4 as we'll be generating next 4 words in the sequence. 
    for i in range(4):
        seq = np.array(tokenizer.texts_to_sequences([input_text])[0]) 
        pred_seq = model.predict_classes(seq)
        
        for word, index in tokenizer.word_index.items():
            if index == pred_seq:
                out_word = word
                break
        input_text = out_word
        result += " "+out_word
    return result
```

```python
generate_text("johny")
```
